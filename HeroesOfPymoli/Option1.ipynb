```python
import pandas as pd

load_file = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(load_file)
purchase_data.head()
```

```python
unique_players = purchase_data['SN'].nunique()
unique_players_df = pd.DataFrame({"Total Players": [unique_players]})
unique_players_df
```

```python
unique_item = purchase_data['Item ID'].nunique()
avg_price = purchase_data['Price'].mean()
tot_purch = len(purchase_data['Item Name'])
tot_rev = purchase_data['Price'].sum()

purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_item],
                                       "Average Price": [avg_price],
                                       "Number of Purchases": [tot_purch],
                                       "Total Revenue": [tot_rev]})

purchasing_analysis_df['Average Price'] = purchasing_analysis_df['Average Price'].map("${:.2f}".format)
purchasing_analysis_df['Total Revenue'] = purchasing_analysis_df['Total Revenue'].map("${:.2f}".format)

purchasing_analysis_df
```

```python
gender_df = pd.DataFrame(purchase_data)

gender_groups = gender_df.groupby(["Gender"])

gender_cnt = gender_groups['SN'].nunique()
gender_perc = (gender_cnt/unique_players)*100

new_gender_df = pd.DataFrame({"Total Count": gender_cnt,
                              "Percentage of Players": gender_perc})

new_gender_df['Percentage of Players'] = new_gender_df['Percentage of Players'].map("{:.2f}%".format) 

sorted_gender_df = new_gender_df.sort_values(by=['Total Count'], ascending=False)

sorted_gender_df.index.name = None

sorted_gender_df
```

```python
gen_unique_purch = gender_groups['Purchase ID'].nunique()
gen_avg_price = gender_groups['Price'].mean()
gen_tot_rev = gender_groups['Price'].sum()
gen_avg_purch = gen_tot_rev/gender_cnt

gen_purchasing_analysis_df = pd.DataFrame({"Purchase Count": gen_unique_item,
                                           "Average Purchase Price": gen_avg_price,
                                           "Total Purchase Value": gen_tot_rev,
                                           "Avg Total Purchase per Person": gen_avg_purch})

gen_purchasing_analysis_df['Average Purchase Price'] = gen_purchasing_analysis_df['Average Purchase Price'].map("${:,.2f}".format)
gen_purchasing_analysis_df['Total Purchase Value'] = gen_purchasing_analysis_df['Total Purchase Value'].map("${:,.2f}".format)
gen_purchasing_analysis_df['Avg Total Purchase per Person'] = gen_purchasing_analysis_df['Avg Total Purchase per Person'].map("${:,.2f}".format)

gen_purchasing_analysis_df
```

```python
age_df = pd.DataFrame(purchase_data)

age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_df["Age Bracket"] = pd.cut(purchase_data["Age"], age_bins, labels=age_groups)
age_group = age_df.groupby("Age Bracket")

age_total_count = age_group['SN'].nunique()
age_perc_players = (age_total_count/unique_players)*100

new_age_df = pd.DataFrame({"Total Count": age_total_count,
                           "Percentage of Players": age_perc_players})

new_age_df['Percentage of Players'] = new_age_df['Percentage of Players'].map("{:.2f}%".format)
new_age_df.index.name = None

new_age_df
```
